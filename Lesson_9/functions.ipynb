```python
def greet(name, location):
    print("Hello", name)
    print(f"You are from {location}")
    return location, name
```

```python
person_location = greet("Cooper", "NYC")
```

```python
print(f"You are located in {person_location[0]}")
```

```python
def calculate_average(score1, score2):
    average = (score1 + score2) / 2
    return average
```

```python
calculate_average(80, 90)
```

```python
average = calculate_average(80, 90)
```

```python
print(f"Hello {person_location[1]}, you are from {person_location[0]} and your grade is {average}")
```

```python
def check_score(score):
    if score >= 65:
        return "passing"
    else:
        return "failing"
```

```python
cooper_score = check_score(18)
student_a_score = check_score(65)
```

```python
print(f"Cooepr Score: {cooper_score}, Student A Score: {student_a_score}")
```

```python
student_scores = [88, 90, 65, 85]
def display_scores(scores):
    for grade in scores:
        print(grade)
```

```python
display_scores(student_scores)
```
