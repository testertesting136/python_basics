# Loops

## For Loops

```python
for i in "hello":
    print("hello")
```

```python
for i in "hello":
    print(i)
```

```python
h = "hello"
```

```python
for i in h:
    print(i *3)
    print("eek!")
```

```python
range(0, 10)
```

```python
for i in range(0, 11):
    print(i)
```

```python
for i in range(-10, 11):
    print(i)
```

```python
for i in range(0, 100, 10):
    print(i)
```

```python
for i in range(-10, 20, -5):
    print(i)
```

```python
for i in range(10, 0, -1):
    print(i)
print("fire!")
```

```python
for i in range(10):
    print(i)
    for j in range(5):
        print(j)
```

```python
for chars in "Cooper":
    print(chars)
```

# While Loops

```python
a = 0
while a <= 10:
    print(a)
    a += 1
```
